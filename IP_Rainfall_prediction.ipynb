## SPLITTING THE DATA

```python
X = df1.drop('RainTomorrow_Yes', axis = 1)                                                # storing all the independent columns in X
y = df1['RainTomorrow_Yes']                                                               # storing the dependent variable in y
```

```python
from sklearn.model_selection import train_test_split                                      # importing train test split library 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)       # splitting the data into train and test
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape                                     # printing the shape
```

### SCALING THE DATA FOR REQUIRED MODELS

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler().fit(X_train)                                                    # scaling the X_train data
X_train_s = scaler.transform(X_train)
scaler = StandardScaler().fit(X_test)                                                     # scaling the X_test data
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV                      # importing GridSearch and RandomizedSearch libraries
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix                               # importing accuracy and confusion matrix
```

```python
from scipy.stats import randint, uniform                                                  # importing randint and uniform 
```

```python
from sklearn.metrics import roc_curve, auc                                                # importing roc_curve and auc 
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_params = {                                                                             # param_grid for logistic regression
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 500, 1000],
}
gv_lr = GridSearchCV(LogisticRegression(),cv=5,param_grid =lr_params,n_jobs= -1,error_score='raise',scoring='accuracy')  # performing grid Search cv to find best parameters
gv_lr.fit(X_train,y_train)
gv_lr.best_params_, gv_lr.best_score_   
```

```python
lrr = gv_lr.best_estimator_                                                               # loading the best model with best parameters
lrr.fit(X_train,y_train)
print(f'train: accuracy_score = {accuracy_score(y_train,lrr.predict(X_train))}')          # printing training score
print(f'test: accuracy_score = {accuracy_score(y_test,lrr.predict(X_test))}')             # printing test score
```

```python
cm_train = confusion_matrix(y_train, lrr.predict(X_train))                                # finding confusion matrix
cm_test = confusion_matrix(y_test, lrr.predict(X_test))
col = {'train':cm_train,
       'test':cm_test}                                                                     
plt.figure(figsize=(15,4))                                                                 
c = 1
for i,j in col.items():
    plt.subplot(1,3,c)                                                                   
    c = c+1
    sn.heatmap(j, annot=True, fmt='d', cmap='Blues', cbar=False)                          # plotting heatmaps
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(i)                                                                   
    

fpr, tpr, thresholds = roc_curve(y_test, lrr.predict(X_test))                             # finding roc_curve
roc_auc = auc(fpr, tpr)                                                                   # finding the auc
plt.subplot(1,3,3)   
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))      # plotting the roc curve
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

## The first two heatmaps shows us the confusion matrix for Train and Test data repectivly. With this we can see the predicted vs actual data. And the third chart is the ROC curve in which the AUC(Area Under The Curve) indicates how well our model is performing and if the AUC > 0.70 then we can say that it is performing good.
